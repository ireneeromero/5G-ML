## Data distibution

Due to the number of columns contained in the csv, viewing a distribution of all the data together would be very complicated to interpret, that is why it will be divided into groups. 

```python
def plot_boxplot(cols, w,h,tit):
    plt.figure(figsize=(10,7))
    fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(w,h))
    for i, col in enumerate(cols):
        df.boxplot(column=[col], fontsize=10)
        plt.subplot(1, len(cols), i+1)
        plt.suptitle(tit)
    plt.show()
```

### DEVI GROUP 

```python
devi_cols = ['DEVI Battery charge','DEVI Battery temperature','DEVI Device CPU usage','DEVI Device battery current']
plot_boxplot(devi_cols, 15, 6, 'DEVI')
```

```python
rlc_cols = ['RLCRATEU RLC uplink block rate','RLCRATEU RLC uplink throughput']
plot_boxplot(rlc_cols, 10, 6, 'RLCRATEU')
```
